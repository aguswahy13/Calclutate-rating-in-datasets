```python
# Import Package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load The Data
df = pd.read_csv("food_delivery_datasets.csv")
df
```

```python
df.info()
```

```python
#Drop missing value dari resto_id & rating
df_1 = df.dropna(subset=['resto_id', 'rating'])
df_1
```

```python
df_1.info()
```

```python
#Drop column
df_2 = df_1.drop(['order_id', 'voucher_amount', 'date_time', 'eta_seconds', 'food_price', 'cust_id', 'delivery_fee', 'status', 'random_id'], axis=1)
df_2
```

```python
# Count unique values in column resto_id of the dataframe
unique_value = df_2['resto_id'].nunique()
print('Number of unique values in column "resto_id" of the dataframe')
print(unique_value)
```

```python
#Count the mean of rating based on resto_id
rating_group = df_2.groupby(['resto_id']).mean()
rating_group
```

```python
#Count the total order each resto_id
resto_group = df_2.groupby(['resto_id']).size()
resto_group
```

```python
#Add "order_count" column
rating_group['order_count'] = resto_group
rating_group
```

```python
#Sort of "order_count" from the largest to smallest
rating_resto = rating_group.sort_values(['order_count','rating'], ascending=False)
rating_resto
```

```python
# Filter "order count" more than and equal 5
rating_resto.loc[rating_resto['order_count'] >=5]
```
